Relation between release year and revenue

```python
f, ax = plt.subplots(figsize=(15.5, 15.5))

sns.scatterplot(y="release_year", x="revenue",
                hue="catigorie",
                palette="tab10",
                sizes=(2, 16), linewidth=0,
                data=movie, ax=ax)
```

before 2000 there are less film categories and less revenue
after 2000 there are more variety in the categories of films and gain more revenue

Divide years into decades
1960 -> 1969 -> 60
1970 -> 1979  -> 70
1980 -> 1989  -> 80
1990 -> 1999  ->  90
2000 -> 2009    -> 20
2010 -> 2019   ->   21

```python
movie['year_range']=movie['release_year'].apply( lambda year: 60 if(year in range(1960,1970))else 70 if(year in range(1970,1980)) else 80 if(year in range(1980,1990))else 90 if(year in range(1990,2000))else 20 if(year in range(2000,2010))else 21 if(year in range(2010,2020)) else 0)
```

Each category and its count throw each decade

```python
f, ax = plt.subplots(figsize=(35, 20))
sns.despine(f)
sns.despine(f, left=True, top=True)
sns.histplot(
    movie,
    hue="year_range", x="catigorie",
    multiple="stack",
    palette="tab10",edgecolor=".5",
    linewidth=.8,

)
```

in 60's  not all the film categories we see now was available  like war, history and  Tv movies
in 90's the number of films are less than other decades
in 20's  we found all categories are available and a large number of films
from 60's to 90's number of films decreasing throw years
and in 20's the number of films increases too much
and back decreasing by 21's
Action,Drama,Comedy and Horror are the most trending categories throw years
but the most popular is Adventure from a previous graph but the number of Adventure films is too small

divide each year into quarters

```python
movie['year_part']=movie['release_date'].apply(lambda month: str(month).split("/")[0])
movie['year_part']=movie['year_part'].apply(lambda month: "q1" if(int(month) in range(1,4))else "q2" if(int(month) in range(4,7))else "q3" if(int(month) in range(7,10)) else "q4" if (int(month) in range(10,13)) else 0 )
```

Relation between revenue and the quarter it releases in

```python
g = sns.catplot(
    data=movie, kind="bar",
    x="year_part", y="revenue",
    palette="tab10", alpha=1, height=20
)
```

The films that released in second quarter of the year gets more revenue
