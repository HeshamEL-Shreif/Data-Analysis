```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("top250-00-19.csv")
```

```python
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

some players have age of 0 so it must be cleaned

```python
data = data[data['Age'] > 0]
```

Relation between position and age

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=data, kind="bar",
    x="Position", y="Age",
     palette="tab10", alpha=1, height=25,
)
```

Most of the players have age less than 25

position and its count hued by season

```python
f, ax = plt.subplots(figsize=(35, 20))
sns.despine(f)

sns.histplot(
    data,
    x="Position",hue="Season",
    multiple="stack",
    palette="tab10",
    edgecolor="1",
    linewidth=1,

)
```

majority of players play as center forward
there are alot of player changes in 2000-2001 season

Relation between position and transfer fees

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=data, kind="bar",
    x="Position", y="Transfer_fee",
    palette="tab10", alpha=1, height=25,
)
```

Transfer fees of Left winger is the highest

Relation between League to and ages

```python
f, ax = plt.subplots(figsize=(10, 25))
sns.despine(f)
sns.boxplot( y="League_to", x="Age", palette="tab10",
            data=data)
```

the most famous leagues buy players in average age 25

Relation between transfer fees and Age

```python


f, ax = plt.subplots(figsize=(10.5, 10.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot( x="Transfer_fee", y="Age",
                palette="tab10",
                sizes=(4, 8), linewidth=0,
                data=data)
```

The highest fees paid for players between 25 and 27 years

Relation between market value and position

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=data, kind="bar",
    x="Position", y="Market_value",
    palette="tab10", alpha=1, height=25,
)
```

left-winger and right-winger have the highest market values

```python
f, ax = plt.subplots(figsize=(10.5, 10.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot( x="Market_value", y="Age",
                 palette="tab10",
                 sizes=(4, 8), linewidth=0,
                 data=data)
```

Ages between 22 and 27 have the highest market values
